# Inventory Management System

```python
import json
```

```python
fd = open('record.json','r')
txt = fd.read()
fd.close()
```

```python
record = json.loads(txt)
```

```python
sales ={}
```

```python
i = 0
n = int(input('Enter the total number of items: '))
for i in range(n):
    p_code = input('Enter the product code:')
    sales.update({i : {'product':record[p_code]['product'], 'brand_name':record[p_code]['brand_name'], 'product_weight':record[p_code]['product_weight'], 'category':record[p_code]['category'], 'price':record[p_code]['price']}} ) 
    i+=1
```

```python
js = json.dumps(sales)
```

```python
fd= open("sales_file.json","w")
fd.write(js)
fd.close()    
```
